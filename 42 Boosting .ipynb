```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#mtcars

df= pd.read_csv("datasets/mtcars.csv")
```

```python
df
```

#####SEPERATE tHE I/O variables


```python
#disp,hp,wt ----Input
x= df[["disp","hp","wt"]]
#mpg ------output
y= df["mpg"]
```

```python
x
```

```python
y
```

```python
x.corrwith(y)    #to find the Corelation x with y
```

##### Build The Model

```python
from sklearn.linear_model import LinearRegression

#create the object
regressor =  LinearRegression()

regressor.fit(x,y)
```

```python
regressor.score(x,y)
```

```python
regressor.coef_      #all three are negatively connected
```

```python
regressor.intercept_
```

##### Predict on unseen data

```python
disp=221
hp=102
wt=2.91


```

```python
new = pd.DataFrame({
    "disp":[221],
    "hp":[102],
    "wt":[2.91]})
```

```python
new
```

```python
regressor.predict(new)           #predicted mpg given on new data.
```

```python
disp * regressor.coef_[0] + hp *regressor.coef_[1] + wt *regressor.coef_[2] + regressor.intercept_             #using farmula
```

```python

plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.scatter(x["disp"],y)
plt.title("Disp vs Mpg")

plt.subplot(1,3,2)
plt.scatter(x["hp"],y)
plt.title("Hp vs Mpg")

plt.subplot(1,3,3)
plt.scatter(x["wt"],y)
plt.title("Wt vs Mpg")
plt.legend()
```

##### ADABoost The Regressor

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
reg_adb =  AdaBoostRegressor(regressor, n_estimators = 3 ,random_state = 0)
```

```python
reg_adb.fit(x,y)
```

```python
reg_adb.score(x,y)
```
