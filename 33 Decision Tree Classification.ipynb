##### Separate the datasets

```python
#inpute Feature

x = df[["Variance","Skewness","Curtosis","Entropy"]]

#output label

y = df["Class"]
```

##### Cross validation

```python
from sklearn.model_selection import train_test_split

x_train ,x_test,y_train ,y_test = train_test_split(x ,y,random_state = 0)
```

```python
x_train     
```

```python
x_test
```

```python
x_train.shape
```

```python
x_test.shape
```

##### Build the model

```python
from sklearn.tree import DecisionTreeClassifier

#created object
clf_treer =  DecisionTreeClassifier(random_state=0)
```

```python
#train the algorithm

clf_treer.fit(x_train,y_train)
```

##### Performance Evaluation

```python
y_pred = clf_treer.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred) 
```

```python
#to print classification report.

print(classification_report(y_test,y_pred))
```

##### Visualize the Tree

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf_treer)
```

```python
plt.figure(figsize=(16,12))             #For Clearity
plot_tree(clf_treer , fontsize = 8 , feature_names=x.columns,filled = True ,precision =2 , class_names = ["O","F"],rounded = True);

plt.savefig("tree.png")                 #To save picture in Current working dir.
```

```python
plt.title("Feature Importances")
plt.xlabel("Feature")
plt.ylabel("Importances")
plt.bar(x.columns , clf_treer.feature_importances_, color ="navy")
```
