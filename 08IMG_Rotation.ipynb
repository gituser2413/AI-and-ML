```python
#Rotation Of Matrix
import cv2

#load the img.
image = cv2.imread("images/lenna.png")

#Define the center points and rotation angle.
center= (image.shape[1]//2 , image.shape[0]//2)
angle= 70

#Create the Rotation mtarix.
rotated_matrix = cv2.getRotationMatrix2D(center,angle,1.0)

#Apply rotation using warpAffine.
rotated_image=cv2.warpAffine(image,rotated_matrix,(image.shape[1] , image.shape[0]))

#Shows The Rotation Img.
cv2.imshow("Rotated IMG",rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Rotation Of Matrix  (Center is change (100,100))
import cv2

#load the img.
image = cv2.imread("images/lenna.png")

#Define the center points and rotation angle.
center= (image.shape[1]//2 , image.shape[0]//2)
angle= 70

#Create the Rotation mtarix.
rotated_matrix = cv2.getRotationMatrix2D((100,100),angle,1.0)

#Apply rotation using warpAffine.
rotated_image=cv2.warpAffine(image,rotated_matrix,(image.shape[1] , image.shape[0]))

#Shows The Rotation Img.
cv2.imshow("Rotated IMG",rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
