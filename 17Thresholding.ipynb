```python
import cv2
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("images/lenna.png",0)
```

```python
new_img = img >127
inv_img = img <127
```

```python
new_img
```

```python
plt.figure(figsize =(16,9))
```

```python
plt.subplot(131)
plt.title("Original IMG")
plt.imshow(img , cmap = "gray")
plt.subplot(132)
plt.title("Binary IMG")
plt.imshow(new_img ,cmap = "gray")
plt.subplot(133)
plt.title(" Inv Binary IMG" )
plt.imshow( inv_img , cmap = "gray")
```
