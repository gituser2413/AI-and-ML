##### DivergingColorMap

```python
pit.figure(figsize=(16,9))
pit.subplot(2,4,1)
pit.title("seismic")
pit.imshow(img,cmap = "seismic")
pit.subplot(2,4,2)
pit.title("coolwarm")
pit.imshow(img,cmap = "coolwarm")
pit.subplot(2,4,3)
pit.title("bwr")
pit.imshow(img,cmap = "bwr")
pit.subplot(2,4,4)
pit.title("Spectral")
pit.imshow(img,cmap = "Spectral")
pit.subplot(2,4,5)
pit.title("PiYG")
pit.imshow(img,cmap = "PiYG")
pit.subplot(2,4,6)
pit.title("BrBG")
pit.imshow(img,cmap = "BrBG")
pit.subplot(2,4,7)
pit.title("RdBu")
pit.imshow(img,cmap = "RdBu")
```

##### Cyclic  ColorMap

```python
pit.figure(figsize=(16,9))
pit.subplot(1,4,1)
pit.title("hsv")
pit.imshow(img,cmap = "hsv")
pit.subplot(1,4,2)
pit.title("turbo")
pit.imshow(img,cmap = "turbo")
pit.subplot(1,4,3)
pit.title("twilight")
pit.imshow(img,cmap = "twilight")
pit.subplot(1,4,4)
pit.title("twilight_shifted")
pit.imshow(img,cmap = "twilight_shifted")
```
