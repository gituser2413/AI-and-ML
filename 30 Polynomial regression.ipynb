##### Polynomial Regression

```python
df= pd.read_csv("datasets/Salary_Data.csv")
x=df[["YearsExperience"]]
y=df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression     #first finding Linear Regression Accuracy

#create the object
reg =  LinearRegression()
reg.fit(x,y)
print("Accuracy",reg.score(x,y))
```

```python
poly = PolynomialFeatures(degree = 2)               # for 2nd degree polynomial accuracy.
x_poly = poly.fit_transform(x)
```

```python
x_poly
```

```python
reg.fit(x_poly,y)
print("Accuracy",reg.score(x_poly,y))
```

```python
poly = PolynomialFeatures(degree = 4)               # for 4th degree polynomial accuracy.
x_poly = poly.fit_transform(x)
x_poly
```

```python
reg.fit(x_poly,y)
print("Accuracy for 4th Degree",reg.score(x_poly,y))
```

```python
acc=[]
for i in range(2,31):
    poly = PolynomialFeatures(degree = i)               # for i th degree from 2 to 20 degree polynomial accuracy.
    x_poly = poly.fit_transform(x)
    reg.fit(x_poly,y)
    acc.append(reg.score(x_poly,y))
   
    
```

```python
acc
```

```python

plt.figure(figsize=(16,9))
plt.grid()
plt.xticks(range(2,31))
plt.xlabel("Degree")
plt.ylabel("Accuracy")
plt.plot(range(2,31), acc, color= "r" ,lw=2 ,marker= "o")
```

```python
poly = PolynomialFeatures(degree = 12)               # for 12  degree from 0 to 12 degree polynomial accuracy.
x_poly = poly.fit_transform(x)
reg.fit(x_poly,y)
reg.score(x_poly,y)
   
    
```

```python
y_pred = reg.predict(x_poly)
```

```python
plt.title("Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.grid()
plt.scatter(x,y,color = "orange", label = "actual")
plt.plot(x,y_pred,color= "blue",label = "predicted")
plt.legend()
```
