```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/wisc_bc_data.csv" )
df.shape
```

```python
 x= df.drop(["id","diagnosis"] , axis = 1)    # two columns are droped id, diagnosis
```

```python
x.shape
```

##### Standardize The Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_std = scaler.fit_transform(x)
```

```python
x.shape
```

##### Apply the PCA

```python
#import tha class 

from sklearn.decomposition import PCA
```

```python
#create the object

pca = PCA()
```

```python
#apply the algo
x_pca =pca.fit_transform(x_std)
```

```python
x_pca
```

```python
x_pca.shape
```

```python
var = pca.explained_variance_ratio_
```

```python
var
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.xlabel("Principle component")
plt.ylabel("Variance")
plt.xticks(range(1,31))
plt.bar(range(1,31),var)
```

```python
#create the object         #for only 10

pca = PCA(n_components = 10)

x_pca =pca.fit_transform(x_std)
                        
x_pca.shape                        
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.xlabel("Principle component")
plt.ylabel("Variance")
plt.xticks(range(1,11))
plt.bar(range(1,11),pca.explained_variance_ratio_)
```
