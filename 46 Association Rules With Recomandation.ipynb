```python
dataset = [["Apple","Beer","Rice","Chicken"],
          ["Apple","Beer","Rice"],
          ["Apple","Beer"],
          ["Apple","Pear"],
          ["Milk","Beer","Rice","Chicken"],
          ["Milk","Beer","Rice"],
          ["Milk","Beer"],
          ["Milk","Pear"]]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded =te.fit_transform(dataset)
```

```python
encoded
```

```python
te.columns_
```

```python
df = pd.DataFrame(encoded,columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori , association_rules
```

```python
#find freqent items
freq_itemset = apriori(df ,min_support=0.25 ,use_colnames=True)
```

```python
freq_itemset
```

```python
rules = association_rules(freq_itemset , metric = "confidence" ,min_threshold= 0.5)
```

```python
rules.shape
```

```python
rules
```

```python
rules = rules[["antecedents","consequents","support","confidence"]]
```

```python
rules
```

```python
##### Sort the Rules
```

```python
rules.sort_values(by =["confidence","support"], ascending=False )
```

##### Filter The Rule

```python
nrules = rules[rules["confidence"] >0.5]
```

```python
nrules.shape
```

```python
nrules
```

```python
nrules[nrules["antecedents"] == {"Apple","Beer"}]["consequents"]
```

```python
nrules[nrules["antecedents"] == {"Chicken","Beer"}]["consequents"]
```

##### Export the Rules

```python
nrules.to_csv("rules.csv" , index = False)
```
