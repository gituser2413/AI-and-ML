```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("images/eiphel1.png",0)
img2 = cv2.imread("images/eiphel2.png",0)
```

```python
plt.figure(figsize =(16,9))
plt.subplot(1,2,1)
plt.title(" IMG1")
plt.imshow(img1 , cmap = "gray")
plt.subplot(1,2,2)
plt.title(" IMG2")
plt.imshow(img2 ,cmap = "gray")
```

```python
sift = cv2.SIFT_create()
```

```python
keypoints_1 , descriptor_1 = sift.detectAndCompute(img1,None)
keypoints_2 , descriptor_2 = sift.detectAndCompute(img2,None)

len(keypoints_1 ),len(keypoints_2 )
```

```python
#feature matching

bf=cv2.BFMatcher(cv2.NORM_L1)

matches = bf.match(descriptor_1,descriptor_2)
matches =sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1, keypoints_1, img2 ,keypoints_2, matches[:20], img2 ,flags=2)

plt.imshow(img3)
```
