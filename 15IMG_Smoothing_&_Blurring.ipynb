```python
import cv2
import numpy as np
#load the img.
image = cv2.imread("images/lenna.png")
```

```python
# Define  a 3*3 averaging Kernel
kernel = np.ones((3,3) , np.float32) / 9

#Apply kernel to smooth the img
smoothen_img = cv2.filter2D(image, -1 , kernel)


cv2.imshow("Original",image)
cv2.imshow("Smoothen",smoothen_img)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
# Define  a 7*7 averaging Kernel
kernel = np.ones((7*7) , np.float32) / 55    #(size of kernel increased )

#Apply kernel to smooth the img
smoothen_img = cv2.filter2D(image, -1 , kernel)


cv2.imshow("Original",image)
cv2.imshow("Smoothen",smoothen_img)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

##### Gaussian Kernel

```python
import matplotlib.pyplot as plt

gray = cv2.imread("images/lenna.png",0)

kernel_size = 5    #5*5
kernel = cv2.getGaussianKernel(kernel_size ,1.25)
blurred = cv2.filter2D(gray , -1 , kernel)  #aply filter
```

```python
kernel
```

```python
cv2.imshow("Original Img IN Gray",gray)
cv2.imshow("Blurred Img",blurred)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
