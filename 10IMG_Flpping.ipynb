```python
import cv2


#load the img.
image = cv2.imread("images/lenna.png")


#Flip the Img.
flipped_horizontal = cv2.flip(image,0)
flipped_vertical = cv2.flip(image,1)
flipped_diagonal = cv2.flip(image,-1)

#Shows The Flipped Img.
cv2.imshow("Horizontal Flip",flipped_horizontal)
cv2.imshow("Vertical Flip",flipped_vertical)
cv2.imshow("Diagonal Flip",flipped_diagonal)
cv2.imshow("Original Img",image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
flip_ho = cv2.cvtColor(flipped_horizontal,cv2.COLOR_BGR2RGB)
flip_ver = cv2.cvtColor(flipped_vertical,cv2.COLOR_BGR2RGB)
flip_dia = cv2.cvtColor(flipped_diagonal,cv2.COLOR_BGR2RGB)
flip_org = cv2.cvtColor(image ,cv2.COLOR_BGR2RGB)
```

```python
#using plt.


import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title("Original IMG")
plt.imshow(flip_org)
plt.subplot(1,4,2)
plt.title("Horizontal Flip")
plt.imshow(flip_ho )
plt.subplot(1,4,3)
plt.title("Diagonal Flip")
plt.imshow(flip_dia)
plt.subplot(1,4,4)
plt.title("Vertical Flip")
plt.imshow(flip_ver)
```
