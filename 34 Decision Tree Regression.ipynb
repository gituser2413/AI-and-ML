```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv("datasets/petrol_consumption.csv")
```

```python
df
```

```python
df.shape
```

##### Separate the I/O data

```python
#inpute Feature

x = df.drop("Petrol_Consumption", axis = 1)

#output label

y = df["Petrol_Consumption"]
```

```python
x.corrwith(y)           #checking the Linearity if 80% matches then use
```

```python
sns.kdeplot(y)        #
```

```python
x.describe()
```

##### Cross Validation

```python
from sklearn.model_selection import train_test_split

x_train ,x_test,y_train ,y_test = train_test_split(x ,y,random_state = 0)
```

```python
x_train.shape ,x_test.shape
```

##### Build The Model

```python
from sklearn.tree import DecisionTreeRegressor

#created object
regressor =  DecisionTreeRegressor(random_state=0)
```

```python
regressor.fit(x_train ,y_train )
```

##### Performance Evaluation

```python
y_pred = regressor.predict(x_test)
```

```python
result = pd.DataFrame({"Actual": y_test , "Predicted": y_pred})
result
```

```python
#Mean Absolute Error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
```

```python
mean_absolute_error( y_test,y_pred)
```

```python
mean_absolute_percentage_error( y_test,y_pred)              #13% error i.e. 87% accuracy
```

```python
np.sqrt(mean_squared_error( y_test,y_pred))
```

##### Visualize

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(16,12))             #For Clearity
plot_tree(regressor , fontsize = 8 , feature_names=x.columns,filled = True ,precision =2 ,rounded = True);

plt.savefig("tree.png")                 #To save picture in Current working dir.
```

```python
plt.title("Feature Importances")
plt.xlabel("Feature")
plt.ylabel("Importances")
plt.bar(x.columns , regressor.feature_importances_, color ="pink")
```
